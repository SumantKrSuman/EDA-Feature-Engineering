# Pie Chart of top 3 Countries

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})

ratings
```

# Observations:- 
1. Rating b/w 4.5 to 4.9- Excellent
2. Rating b/w 4.0 to 4.4- Very Good
3. Rating b/w 3.5 to 3.9- Good
4. Rating b/w 2.5 to 3.4- Average
5. Rating b/w 1.8 to 2.4- Bad
6. Approx 2K people have not Rated the Experience

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x= 'Aggregate rating', y= 'Rating count', data= ratings)
```

```python
# sns.barplot(x= 'Aggregate rating', y= 'Rating count', hue= 'Rating color', data= ratings)

sns.barplot(x= 'Aggregate rating', y= 'Rating count', hue= 'Rating color', data= ratings,palette= ['Black', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```

# Observation:-
    1. Not Rated Count is very high.
    2. Maximum number of Rating is b/w 2.5 to 4.0

# Count plot for frequency of Rating

```python
sns.countplot(x= 'Rating color', data= ratings,palette= ['Black', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```
