### Feature Engineering with Route:
    Either we can split Route in 4 Parts but that will create empty values for those flights which have less than 4 Stops.
    So we will drop Route and work with Total_Stops

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']= final_df['Total_Stops'].map({'non-stop':0,
                                                     '1 stop':1,
                                                     '2 stops':2,
                                                     '3 stops':3,
                                                     '4 stops':4,
                                                     'nan':1})
```

```python
final_df.head()
```

```python
final_df.drop('Route', axis= 1, inplace= True)
```

```python
final_df.head(1)
```

### Feature Engineering with Duration

###### Extracting Hour from Duration Column.

```python
final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df['Duration_hour']= final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.head(1)
```

```python
final_df[final_df['Duration_hour']=='5m']
```

```python
final_df.drop(6474,axis= 0, inplace= True)
```

```python
final_df.drop(2660,axis= 0, inplace= True)
```

We should drop these two records as it is impossible to have only 5minute flight with 2 stops

```python
final_df['Duration_hour']= final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_hour']= final_df['Duration_hour']*60
```

#### Extracting Minute from Duration Column.

```python
final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration_min']= final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.head(1)
```

```python
final_df['Duration_min'].unique()
```

```python
final_df[final_df['Duration_min'].isnull()]
```

```python
final_df['Duration_min'] = final_df['Duration_min'].fillna(0)
```

```python
final_df[final_df['Duration_min'].isnull()]
```

```python
final_df['Duration_min']= final_df['Duration_min'].astype(int)
```

```python
final_df.drop('Duration', axis= 1, inplace= True)
```

```python
final_df.head(1)
```

```python
# Adding Duration hour and Duration Minute

final_df['Duration_min']= final_df['Duration_hour']+final_df['Duration_min']
```

```python
final_df.head(1)
```

```python
# Drop Duration Hour as we do not need it now

final_df.drop('Duration_hour', axis= 1, inplace= True)
```

```python
final_df.head(1)
```

### Feature Engineering with Categorical Features

```python
final_df['Airline'].unique()
```

```python
# We will do Label Encoding here as dummy variable will lead to more memory consumption

from sklearn.preprocessing import LabelEncoder

labelencoder= LabelEncoder()
```

```python
final_df['Airline']= labelencoder.fit_transform(final_df['Airline'])
final_df['Source']= labelencoder.fit_transform(final_df['Source'])
final_df['Destination']= labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']= labelencoder.fit_transform(final_df['Additional_Info'])
```
